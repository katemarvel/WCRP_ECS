```python
%matplotlib notebook
import scipy.stats as stata
import numpy as np
import matplotlib.pyplot as plt
nsamp=100

def sigma(sf,sl):
    test=[]
    for i in range(nsamp):
        F=np.random.normal(3.7,sf,size=100)
        L=np.random.normal(1.1,sl,size=100)
        test+=[np.var(F/L)]
    return np.average(np.array(test))

```

```python
testf=[sigma(sf,0) for sf in np.linspace(0,1,.1)]
testl=[sigma(0,sl) for sl in np.linspace(0,1,.1)]
```

```python
plt.plot(testf)
plt.plot(testl)
```
